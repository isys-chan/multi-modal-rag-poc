# chroma basic

```python
# chroma setup
import chromadb
chroma_client = chromadb.PersistentClient(path="./chroma_db") # PersistentClientにすると、DBがfolder保存される
```

```python
# OpenAI enbedding setup
import numpy as np
import os
from chromadb.utils import embedding_functions
from dotenv import load_dotenv
load_dotenv()

openai_ef = embedding_functions.OpenAIEmbeddingFunction(
    api_key = os.environ["OPENAI_API_KEY"],
    model_name="text-embedding-3-small"
    )
```

```python
 # コレクション= テーブルの作成。Chromaフォルダ内でUUIDで識別される
collection_name = "yes_collection"
collection = chroma_client.create_collection(
    name=collection_name, 
    embedding_function=openai_ef, # embedding_functionを指定。ここで指定するとADD時に自動でembeddingされる
    metadata={"hnsw:space":"cosine"} # コサイン類似度
    )

```

```python
# adding document to collection
collection.add (
    documents=["This is a sample document."], # ドキュメント内容
    metadatas=[{"source": "sample_source"}], # なんかIDとか好きに入れれる、書類の種類とか
    ids=["doc1"]# chunk_idは一意のidになるように
    )
```

```python
# seraching collection
query = "sample"
result = collection.query(
    query_texts=[query], # 検索クエリ(query_textは一致検索, query_embeddingsはベクトル検索をする。どちらかを指定可能)
    n_results=3 # 返す件数
    # where={} でメタデータ絞り込みも可能
    )
print(result)
print("---------------")
print(result["documents"][0][0])
```
